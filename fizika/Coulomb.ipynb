## Coulomb-törvény

Legyen $\mathbf{r}$ a két töltés közötti vektor, $q_1$ és $q_2$ a két töltés nagyága, $k$ pedig a Coulomb-féle arányossági tényező, amelyet most az egyszerűség kedvéért $1$-nek veszünk. A Coulomb-törvény alapján a két töltés között fellépő erő
$$ \mathbf{F} = k \cdot \frac{q_1 q_2}{\left|\mathbf{r}\right|^3} \mathbf{r} .$$
Dolgozzunk a síkban, helyezzünk el néhány $\pm 1$ töltést tetszőleges koordinátákra, majd számoljuk ki a sík minden egyes pontjában a fellépő erők eredőjét (azaz a sík egy adott pontjában számoljuk ki az $\mathbf{F}$ vektormennyiséget mindegyik töltésre nézve, majd összegezzük őket). Az erővonalakat `streamplot` segítségével ábrázoljuk.

```python
%pylab inline
```

```python
def F(q,r,x,y):
    dx,dy = x-r[0],y-r[1]
    d = np.sqrt(dx*dx+dy*dy)**3
    return q*dx/d,q*dy/d

p   = 256
x   = np.linspace(-4,4,p)
y   = np.linspace(-4,4,p)
X,Y = np.meshgrid(x,y)

q   = [1,-1,-1]
pos = [(-1,0),(1,0),(2,2)]

Fx,Fy = np.zeros((p,p)), np.zeros((p,p))
for i in range(len(q)):
    fx,fy = F(q[i],pos[i],X,Y)
    Fx += fx
    Fy += fy
```

```python
streamplot(x,y,Fx,Fy,density=3,color=np.log(Fx**2+Fy**2),cmap="hot",arrowstyle="->");
```
