Mandelbrot-halmaz:

https://hu.wikipedia.org/wiki/Mandelbrot-halmaz

Magyarázat a kódhoz:

- `np.add.outer()` két vektort ad össze, de nem elemenként, hanem mindegyik elemet mindegyik elemmel, egy mátrixba elrendezve,
- `np.linspace(-2,2,n)` a $[-2,2]$ intervallumba eső számok közül tartalmaz $n$ darabot egyenletesen elosztva,
- `1j` az imaginárius egység ($i$),
- `np.flatten()` "kilapítja" a mátrixot, azaz a sorait egymás után rakja (így könnyebb lesz az indexelés).

A `c` változó tehát a komplex számsík $[-2,2] \times [-2i,2i]$ részét tartalmazza kilapítva. Ezután a $z \gets z^2+c$ iterációt alkalmazzuk, és megnézzük, hogy hol "száll el" a kapott sorozat (pontosabban hány iteráción keresztül marad $2$ alatt az abszolútérték egy adott pontból indítva, ezek kerülnek az `n` változóba). Végül `n`-t visszarendezzük mátrixszá és ábrázoljuk ($\sim$ hol milyen gyorsan száll el a sorozat).

```python
%pylab inline
```

```python
x,y = 1000,1000

c = np.add.outer(np.linspace(-2,2,x),np.linspace(-2,2,y)*1j).flatten()
z = c.copy()
n = np.zeros(x*y)
for i in range(100):
  z    = z**2 + c
  
  # Ezeket is próbáld ki:
  #z    = (z**2+c)**2/z
  #z    = np.sinh(z+c)
  e    = np.abs(z)<2
  n[e] = i
figsize(10,10)
imshow(n.reshape(x,y),cmap='hot');
```
